```python
import pandas as pd
from sqlalchemy import create_engine
from config_mg import passwd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_file = "./Crime_ca.csv"
crime_data_df = pd.read_csv(csv_file,encoding = "ISO-8859-1")
crime_data_df.head(100)
```

```python
crime_data_df.columns
```

```python
clean_crime_data = crime_data_df[['reportyear','geotype','geoname','county_name','region_name','strata_level_name','numerator', 'denominator', 'rate']]
CA_crime_data = clean_crime_data
CA_crime_data
```

```python
CA_crime_data = CA_crime_data.rename(columns={"numerator": "no_of_crime", "denominator":"population"})
CA_crime_data['rate'] = (CA_crime_data['no_of_crime']/CA_crime_data['population'])*1000
CA_crime_data.reset_index()   
CA_crime_data.index.name = 'id'
len(CA_crime_data)
```

```python
CA_crime_data.columns
#CA_crime_data.index.dtype
```

```python
total_crime_rate= CA_crime_data.loc[CA_crime_data['strata_level_name']== 'Violent crime total']
year_crime = total_crime_rate.groupby(['reportyear']).sum()
year_crime.reset_index(inplace = True)
year_crime
```

```python
year_crime.plot(kind="bar", x='reportyear', y='rate', grid=True, figsize=(8,8),
              title="rate vs year")
plt.show()
```

```python
plt.savefig("./images_analysis/crime_year_bar.png")
```

```python
region_crime = total_crime_rate.groupby(['county_name']).sum()
region_crime.reset_index(inplace = True)
```

```python
region_crime.plot(kind="bar", x='county_name', y='rate', grid=True, figsize=(8,8),
              title="rate vs region")
plt.show()
```

```python
plt.savefig("./images_analysis/crime_region_bar.png")
```
